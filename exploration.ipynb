```python
import requests as req 
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
CTS_TOKEN=os.getenv("CTS_TOKEN")
BASE_URL="https://api.cts-strasbourg.eu"
AUTH=(CTS_TOKEN,"")
```

```python
url=BASE_URL+"/v1/siri/2.0/estimated-timetable"

params = {
    "LineRef": "CTS:Line:A",
    "StopPointRef": "CTS:StopPoint:HOH",
    "MaximumStopVisits": 3
}

resp=req.get(
    url,
    params=params,
    auth=AUTH)
```
